###  Описание данных


## Задание 1



1. Для столбца tr_type датафрейма transactions выберите произвольные 1000 строк с помощью метода sample
2. В полученной на предыдущем этапе подвыборке найдите долю транзакций (стобец tr_description в датасете tr_types), в которой содержится подстрока 'POS' или 'ATM'

````python
transactions_sample = transactions['tr_type'].sample(n=1000, random_state=23)
````

````python
transactions_sample.reset_index(drop=True, inplace=True)
````

````python
grouped_trans = transactions_sample.groupby(transactions_sample.values).count()
````

````python
sought_trans = 0
for i in list(grouped_trans.index):
    descr = tr_types[(tr_types['tr_type'] == i)]['tr_description'].values
    if not descr:
        continue
    print(descr)
    if 'POS' in descr[0] or 'АТМ' in descr[0]:
        sought_trans += 1 * grouped_trans[i]
        print(sought_trans, grouped_trans[i])
````

````python
share_of_trans = sought_trans / 1000
print(f'Доля транзакций: {share_of_trans}')
````

## Задание 8

1. Из поля tr_datetime выделите час tr_hour, в который произошла транзакция, как первые 2 цифры до ":". (\**)
2. Посчитайте ночные поступления для мужчин и женщин. Ночные - это примерно 00-06 часов. 

````python
def hour_parser(datetime_str):
    for i in range(len(datetime_str)):
        if datetime_str[i] == ' ':
            h = datetime_str[i+1:i+3]
            break
    return int(h)
````

````python
tr_hour = [hour_parser(transactions['tr_datetime'][i]) for i in range(len(transactions))]
````

````python
transactions['tr_hour'] = tr_hour
````

````python
transactions.head()
````

````python
night_arrivals_0 = transactions[(transactions["gender"] == 0)]['amount'] \
                    .where((transactions["amount"] > 0) & (transactions["tr_hour"].between(0, 6, inclusive=False))).dropna().sum()

night_arrivals_1 = transactions[(transactions["gender"] == 1)]['amount'] \
                    .where((transactions["amount"] > 0) & (transactions["tr_hour"].between(0, 6, inclusive=False))).dropna().sum()
````

````python
night_arrivals_0
````
